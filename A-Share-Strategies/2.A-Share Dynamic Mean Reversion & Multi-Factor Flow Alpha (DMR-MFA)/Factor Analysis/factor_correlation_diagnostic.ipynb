# Factor Correlation Diagnostic`n`n## Research Objective`nThis notebook evaluates the correlation structure, overlap intensity, and diversification quality of the full factor set used in the Dynamic Mean Reversion & Multi-Factor Flow Alpha framework. The goal is to determine whether the factor library provides sufficiently differentiated information, or whether some factors are mainly duplicating the same short-term return driver.`n

## Factor Definition`nThe diagnostic covers the strategy's main stock-level factors and separates them from the benchmark-level market regime factor:`n`n- **Stock-level cross-sectional factors**`n  - BBI Slope`n  - OBV Divergence`n  - 10-Day Downtrend`n  - Net Main Inflow`n  - 3-Day Drawdown`n- **Benchmark-level context factor**`n  - HS300 Bollinger Regime`n`n## Research Hypothesis`nThe working hypothesis is that a healthy multi-factor framework should contain both related and independent signals. Some moderate overlap is expected, especially among technical factors describing similar short-term price stress, but excessively high correlation would indicate redundancy and reduce the benefit of combining multiple factors in the final strategy logic.`n`n## Data and Test Design`nThe notebook reconstructs the full factor set on the strategy's fixed 33-stock pool over **2016-01-01 to 2026-04-29**, using monthly cross-sectional snapshots aligned with the rest of the factor-analysis framework. The study includes average monthly Pearson and Spearman correlation matrices, high-correlation frequency diagnostics, signal-ratio interaction analysis, strongest-pair scatter inspection, and HS300 market-context linkage analysis.`n

```python
'''
Multi-factor correlation diagnostic for the Dynamic Mean Reversion &
Multi-Factor Flow Alpha framework.

Research purpose:
1. Evaluate cross-sectional correlation among the stock-level factors that
   jointly drive candidate selection or exit filtering.
2. Separate stock-level factor overlap from benchmark-level market-regime
   information, because the HS300 Bollinger regime is a macro context factor
   rather than a pure stock cross-sectional signal.
3. Provide a professional diagnostic view of redundancy, diversification,
   and structural interaction across the full factor set used by the strategy.

Methodology:
1. Use the strategy's fixed 33-stock pool.
2. Sample the last trading day of each month from 2016-01-01 to 2026-04-29.
3. Reconstruct each factor with the same definitions already used in the
   standalone factor-analysis scripts.
4. Build:
   - average monthly Pearson correlation matrix on winsorized score values
   - average monthly Spearman correlation matrix on raw cross-sectional scores
   - high-correlation frequency matrix
   - market-context correlation heatmap between HS300 regime variables and
     monthly stock-factor summary variables
'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from jqdata import *
from scipy.stats import spearmanr
import warnings

warnings.filterwarnings('ignore')
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['axes.unicode_minus'] = False
sns.set_style("whitegrid")

STOCK_POOL = [
    '601117.XSHG', '601600.XSHG', '601888.XSHG', '300274.XSHE', '300750.XSHE',
    '601919.XSHG', '002049.XSHE', '603881.XSHG', '002335.XSHE', '600089.XSHG',
    '002236.XSHE', '002056.XSHE', '300866.XSHE', '002611.XSHE', '600760.XSHG',
    '300693.XSHE', '002402.XSHE', '002600.XSHE', '300207.XSHE', '603486.XSHG',
    '000591.XSHE', '000027.XSHE', '600011.XSHG', '601899.XSHG', '603799.XSHG',
    '002340.XSHE', '002780.XSHE', '600160.XSHG', '601225.XSHG', '002555.XSHE',
    '600803.XSHG', '300059.XSHE', '002736.XSHE',
]

SCORE_COLS = [
    'bbi_slope_score',
    'obv_divergence_score',
    'ten_day_downtrend_score',
    'net_main_inflow_score',
    'three_day_drawdown_score',
]

SIGNAL_COLS = [
    'bbi_bearish_signal',
    'obv_divergence_signal',
    'ten_day_downtrend_signal',
    'net_main_inflow_positive_signal',
    'three_day_drawdown_threshold_signal',
]

FACTOR_LABELS = {
    'bbi_slope_score': 'BBI Slope',
    'obv_divergence_score': 'OBV Divergence',
    'ten_day_downtrend_score': '10D Downtrend',
    'net_main_inflow_score': 'Net Main Inflow',
    'three_day_drawdown_score': '3D Drawdown',
}

SIGNAL_LABELS = {
    'bbi_bearish_signal': 'BBI Bearish',
    'obv_divergence_signal': 'OBV Divergence',
    'ten_day_downtrend_signal': '10D Downtrend',
    'net_main_inflow_positive_signal': 'Main Inflow > 0',
    'three_day_drawdown_threshold_signal': '3D Drawdown < -7%',
}


def calc_bbi(close_array):
    """Builds the BBI proxy used by the strategy."""

    ma10 = close_array[-10:].mean()
    ma20 = close_array[-20:].mean()
    ma30 = close_array[-30:].mean()
    ma60 = close_array[-60:].mean()
    return (ma10 + ma20 + ma30 + ma60) / 4.0


def winsorize_series(series, lower=0.05, upper=0.95):
    """Winsorizes a series to reduce one-month extreme-value distortion."""

    clean = pd.Series(series).copy()
    if clean.dropna().empty:
        return clean
    low = clean.quantile(lower)
    high = clean.quantile(upper)
    return clean.clip(lower=low, upper=high)


def zscore_series(series):
    """Standardizes a series while guarding against zero-variance months."""

    clean = pd.Series(series).astype(float)
    std = clean.std()
    if pd.isna(std) or std == 0:
        return clean * 0.0
    return (clean - clean.mean()) / std


def safe_average_matrix(matrix_list, labels):
    """Averages same-shaped correlation matrices."""

    if len(matrix_list) == 0:
        return pd.DataFrame(np.nan, index=labels, columns=labels)
    arr = np.stack([m.values for m in matrix_list], axis=0)
    return pd.DataFrame(np.nanmean(arr, axis=0), index=labels, columns=labels)


def top_correlation_pairs(corr_df, top_n=10):
    """Extracts the strongest off-diagonal absolute-correlation pairs."""

    pairs = []
    cols = list(corr_df.columns)
    for i in range(len(cols)):
        for j in range(i + 1, len(cols)):
            pairs.append({
                'factor_1': cols[i],
                'factor_2': cols[j],
                'corr': corr_df.iloc[i, j],
                'abs_corr': abs(corr_df.iloc[i, j]),
            })
    pairs = pd.DataFrame(pairs).sort_values('abs_corr', ascending=False)
    return pairs.head(top_n)


def plot_heatmap(ax, data, title, center=0, fmt='.2f', cmap='RdBu_r', annot=True):
    """Draws a consistent heatmap for diagnostic sections."""

    sns.heatmap(data, annot=annot, fmt=fmt, cmap=cmap, center=center, ax=ax)
    ax.set_title(title, fontsize=14)


def build_hs300_context(bench_df):
    """Computes the benchmark Bollinger context variables."""

    closes = bench_df['close'].values
    ma20 = np.mean(closes[:-1])
    std20 = np.std(closes[:-1])
    current = closes[-1]
    upper = ma20 + 2 * std20
    lower = ma20 - 2 * std20

    if current > upper:
        regime = 'up'
        regime_code = 1
        target_num = 7
    elif current < lower:
        regime = 'down'
        regime_code = -1
        target_num = 3
    else:
        regime = 'neutral'
        regime_code = 0
        target_num = 5

    return {
        'hs300_regime': regime,
        'hs300_regime_code': regime_code,
        'hs300_target_num': target_num,
        'hs300_band_width': (upper - lower) / ma20 if ma20 != 0 else np.nan,
        'hs300_band_position': (current - ma20) / std20 if std20 != 0 else 0.0,
    }


# ------------------------------------------------------------------------------
# 1. Sampling Framework
# ------------------------------------------------------------------------------
start_date, end_date = '2016-01-01', '2026-04-29'
all_days = get_trade_days(start_date, end_date)
ts_series = pd.to_datetime(pd.Series(all_days))
rebal_dates = ts_series.groupby([ts_series.dt.year, ts_series.dt.month]).last()
date_list = [d.strftime('%Y-%m-%d') for d in rebal_dates]

dataset = []
print('Initiating Multi-Factor Correlation Diagnostic...')

# ------------------------------------------------------------------------------
# 2. Monthly Factor Reconstruction
# ------------------------------------------------------------------------------
# For each monthly snapshot:
# - rebuild all stock-level factor scores
# - attach the same HS300 Bollinger regime context to each stock row
for i in range(len(date_list) - 1):
    t0, t1 = date_list[i], date_list[i + 1]
    print('\rProcessing Snapshot: {}'.format(t0), end='')

    hist_close = get_price(STOCK_POOL, end_date=t0, count=61, fields=['close'], panel=False, fq='pre')
    hist_obv = get_price(STOCK_POOL, end_date=t0, count=10, fields=['close', 'volume'], panel=False, fq='pre')
    hold = get_price(STOCK_POOL, start_date=t0, end_date=t1, fields=['close'], panel=False, fq='pre')
    money_flow = get_money_flow(STOCK_POOL, start_date=t0, end_date=t0, fields=['sec_code', 'date', 'net_amount_main'])
    bench = get_price('000300.XSHG', end_date=t0, count=21, fields=['close'], fq='pre')

    if (
        hist_close is None or hist_close.empty or
        hist_obv is None or hist_obv.empty or
        hold is None or hold.empty or
        money_flow is None or money_flow.empty or
        bench is None or len(bench) < 21
    ):
        continue

    market_ctx = build_hs300_context(bench)

    for stock in STOCK_POOL:
        h_close = hist_close[hist_close['code'] == stock].sort_values('time').reset_index(drop=True)
        h_obv = hist_obv[hist_obv['code'] == stock].sort_values('time').reset_index(drop=True)
        p = hold[hold['code'] == stock].sort_values('time').reset_index(drop=True)
        mf = money_flow[money_flow['sec_code'] == stock]

        if len(h_close) < 61 or len(h_obv) < 10 or len(p) < 2 or mf.empty:
            continue

        # BBI slope factor
        close_now = h_close['close'].values[-60:]
        close_prev = h_close['close'].values[-61:-1]
        curr_bbi = calc_bbi(close_now)
        prev_bbi = calc_bbi(close_prev)
        bbi_raw = curr_bbi - prev_bbi
        bbi_score = -bbi_raw
        bbi_signal = 1 if curr_bbi < prev_bbi else 0

        # OBV divergence factor
        diff = h_obv['close'].diff()
        direction = diff.apply(lambda x: 1 if x > 0 else (-1 if x < 0 else 0))
        obv = (direction * h_obv['volume']).fillna(0).cumsum()
        obv_prev_high = obv.iloc[-6:-1].max()
        price_prev_high = h_obv['close'].iloc[-6:-1].max()
        obv_raw = obv.iloc[-1] - obv_prev_high
        obv_score = -obv_raw
        obv_signal = 1 if (obv.iloc[-1] > obv_prev_high) and (h_obv['close'].iloc[-1] <= price_prev_high) else 0

        # 10-day downtrend factor
        ten_raw = h_close['close'].iloc[-1] / h_close['close'].iloc[-11] - 1.0
        ten_score = -ten_raw
        ten_signal = 1 if h_close['close'].iloc[-1] < h_close['close'].iloc[-11] else 0

        # Net main inflow factor
        net_raw = float(mf['net_amount_main'].iloc[-1])
        if np.isnan(net_raw):
            continue
        net_score = net_raw
        net_signal = 1 if net_raw > 0 else 0

        # 3-day drawdown factor
        draw_raw = h_close['close'].iloc[-1] / h_close['close'].iloc[-4] - 1.0
        draw_score = -draw_raw
        draw_signal = 1 if draw_raw < -0.07 else 0

        fwd_ret = p['close'].iloc[-1] / p['close'].iloc[0] - 1.0

        dataset.append({
            'date': t0,
            'code': stock,
            'bbi_slope_raw': bbi_raw,
            'bbi_slope_score': bbi_score,
            'bbi_bearish_signal': bbi_signal,
            'obv_divergence_raw': obv_raw,
            'obv_divergence_score': obv_score,
            'obv_divergence_signal': obv_signal,
            'ten_day_downtrend_raw': ten_raw,
            'ten_day_downtrend_score': ten_score,
            'ten_day_downtrend_signal': ten_signal,
            'net_main_inflow_raw': net_raw,
            'net_main_inflow_score': net_score,
            'net_main_inflow_positive_signal': net_signal,
            'three_day_drawdown_raw': draw_raw,
            'three_day_drawdown_score': draw_score,
            'three_day_drawdown_threshold_signal': draw_signal,
            'fwd_ret': fwd_ret,
            'hs300_regime': market_ctx['hs300_regime'],
            'hs300_regime_code': market_ctx['hs300_regime_code'],
            'hs300_target_num': market_ctx['hs300_target_num'],
            'hs300_band_width': market_ctx['hs300_band_width'],
            'hs300_band_position': market_ctx['hs300_band_position'],
        })

print('\nData collection complete.')

# ------------------------------------------------------------------------------
# 3. Master Dataset Preparation
# ------------------------------------------------------------------------------
master_df = pd.DataFrame(dataset)
master_df['date'] = pd.to_datetime(master_df['date'])
master_df['year'] = master_df['date'].dt.year
master_df['month'] = master_df['date'].dt.month

# ------------------------------------------------------------------------------
# 4. Cross-Sectional Correlation Diagnostics
# ------------------------------------------------------------------------------
pearson_mats = []
spearman_mats = []
high_corr_mats = []
coverage_rows = []
standardized_rows = []

for date, group in master_df.groupby('date'):
    sub = group[SCORE_COLS].copy()
    sub = sub.replace([np.inf, -np.inf], np.nan).dropna()
    if len(sub) < 8:
        continue

    winsorized = sub.copy()
    for col in SCORE_COLS:
        winsorized[col] = winsorize_series(winsorized[col])
        winsorized[col] = zscore_series(winsorized[col])

    pearson_mat = winsorized.corr(method='pearson')
    spearman_mat = sub.corr(method='spearman')
    high_corr_flag = (spearman_mat.abs() >= 0.6).astype(float)
    np.fill_diagonal(high_corr_flag.values, 0.0)

    pearson_mats.append(pearson_mat)
    spearman_mats.append(spearman_mat)
    high_corr_mats.append(high_corr_flag)

    coverage_rows.append({
        'date': date,
        'sample_size': len(sub),
        'mean_abs_pearson': pearson_mat.where(~np.eye(len(SCORE_COLS), dtype=bool)).abs().stack().mean(),
        'mean_abs_spearman': spearman_mat.where(~np.eye(len(SCORE_COLS), dtype=bool)).abs().stack().mean(),
    })

    standardized = winsorized.copy()
    standardized['date'] = date
    standardized_rows.append(standardized)

avg_pearson_raw = safe_average_matrix(pearson_mats, SCORE_COLS)
avg_spearman_raw = safe_average_matrix(spearman_mats, SCORE_COLS)
high_corr_freq_raw = safe_average_matrix(high_corr_mats, SCORE_COLS)
coverage_df = pd.DataFrame(coverage_rows).set_index('date')
stacked_df = pd.concat(standardized_rows, axis=0).reset_index(drop=True) if len(standardized_rows) > 0 else pd.DataFrame()

top_pairs_raw = top_correlation_pairs(avg_spearman_raw, top_n=10)

avg_pearson = avg_pearson_raw.rename(index=FACTOR_LABELS, columns=FACTOR_LABELS)
avg_spearman = avg_spearman_raw.rename(index=FACTOR_LABELS, columns=FACTOR_LABELS)
high_corr_freq = high_corr_freq_raw.rename(index=FACTOR_LABELS, columns=FACTOR_LABELS)

# ------------------------------------------------------------------------------
# 5. Market-Regime Interaction Diagnostics
# ------------------------------------------------------------------------------
monthly_summary = master_df.groupby('date').agg({
    'bbi_slope_score': 'mean',
    'obv_divergence_score': 'mean',
    'ten_day_downtrend_score': 'mean',
    'net_main_inflow_score': 'mean',
    'three_day_drawdown_score': 'mean',
    'bbi_bearish_signal': 'mean',
    'obv_divergence_signal': 'mean',
    'ten_day_downtrend_signal': 'mean',
    'net_main_inflow_positive_signal': 'mean',
    'three_day_drawdown_threshold_signal': 'mean',
    'fwd_ret': 'mean',
    'hs300_regime_code': 'first',
    'hs300_target_num': 'first',
    'hs300_band_width': 'first',
    'hs300_band_position': 'first',
})

market_cols = [
    'hs300_regime_code',
    'hs300_target_num',
    'hs300_band_width',
    'hs300_band_position',
]

factor_monthly_cols = [
    'bbi_slope_score',
    'obv_divergence_score',
    'ten_day_downtrend_score',
    'net_main_inflow_score',
    'three_day_drawdown_score',
    'bbi_bearish_signal',
    'obv_divergence_signal',
    'ten_day_downtrend_signal',
    'net_main_inflow_positive_signal',
    'three_day_drawdown_threshold_signal',
    'fwd_ret',
]

market_factor_corr = monthly_summary[market_cols + factor_monthly_cols].corr().loc[market_cols, factor_monthly_cols]
market_factor_corr = market_factor_corr.rename(columns={
    'bbi_slope_score': 'BBI Mean Score',
    'obv_divergence_score': 'OBV Mean Score',
    'ten_day_downtrend_score': '10D Mean Score',
    'net_main_inflow_score': 'Main Inflow Mean Score',
    'three_day_drawdown_score': '3D Mean Score',
    'bbi_bearish_signal': 'BBI Signal Ratio',
    'obv_divergence_signal': 'OBV Signal Ratio',
    'ten_day_downtrend_signal': '10D Signal Ratio',
    'net_main_inflow_positive_signal': 'Main Inflow Positive Ratio',
    'three_day_drawdown_threshold_signal': '3D Threshold Ratio',
    'fwd_ret': 'Pool Mean Forward Return',
}).rename(index={
    'hs300_regime_code': 'HS300 Regime Code',
    'hs300_target_num': 'HS300 Target Num',
    'hs300_band_width': 'HS300 Band Width',
    'hs300_band_position': 'HS300 Band Position',
})

signal_ratio_df = monthly_summary[SIGNAL_COLS].rename(columns=SIGNAL_LABELS)
signal_corr = signal_ratio_df.corr()

# ------------------------------------------------------------------------------
# 6. Visualization: Part I - Stock-Level Correlation Structure
# ------------------------------------------------------------------------------
fig1 = plt.figure(figsize=(24, 22))
plt.suptitle('Factor Correlation Diagnostic: Part I', fontsize=24, fontweight='bold', y=0.96)

ax1 = plt.subplot(3, 2, 1)
plot_heatmap(ax1, avg_pearson, 'I. Average Monthly Pearson Correlation', fmt='.2f')

ax2 = plt.subplot(3, 2, 2)
plot_heatmap(ax2, avg_spearman, 'II. Average Monthly Spearman Correlation', fmt='.2f')

ax3 = plt.subplot(3, 2, 3)
plot_heatmap(ax3, high_corr_freq, 'III. Frequency of |Spearman Corr| >= 0.60', fmt='.0%')

ax4 = plt.subplot(3, 2, 4)
mean_abs_bar = avg_spearman.abs().replace(1.0, np.nan).mean(axis=1).sort_values(ascending=False)
mean_abs_bar.plot(kind='bar', color='#457b9d', ax=ax4)
ax4.set_title('IV. Average Absolute Spearman Correlation by Factor', fontsize=14)
ax4.set_ylabel('Average |Correlation|')

ax5 = plt.subplot(3, 2, 5)
coverage_df['sample_size'].plot(ax=ax5, color='#2a9d8f', lw=2)
ax5.set_title('V. Monthly Common Sample Size', fontsize=14)
ax5.set_ylabel('Stock Count')

ax6 = plt.subplot(3, 2, 6)
coverage_df[['mean_abs_pearson', 'mean_abs_spearman']].plot(ax=ax6, lw=2, color=['#e76f51', '#1d3557'])
ax6.set_title('VI. Monthly Mean Absolute Correlation Intensity', fontsize=14)
ax6.axhline(0, color='black', lw=1)

plt.tight_layout(rect=[0, 0.03, 1, 0.93])
plt.show()

# ------------------------------------------------------------------------------
# 7. Visualization: Part II - Market Context and Signal Interaction
# ------------------------------------------------------------------------------
fig2 = plt.figure(figsize=(24, 24))
plt.suptitle('Factor Correlation Diagnostic: Part II', fontsize=24, fontweight='bold', y=0.96)

ax7 = plt.subplot(3, 2, 1)
plot_heatmap(ax7, market_factor_corr, 'VII. HS300 Context vs Monthly Factor Summary', fmt='.2f')

ax8 = plt.subplot(3, 2, 2)
signal_ratio_df.plot(ax=ax8, lw=2)
ax8.set_title('VIII. Monthly Signal-Ratio Time Series', fontsize=14)
ax8.set_ylabel('Signal Ratio')
ax8.legend(loc='best', fontsize=8)

ax9 = plt.subplot(3, 2, 3)
annual_signal = signal_ratio_df.copy()
annual_signal['year'] = annual_signal.index.year
annual_signal.groupby('year').mean().plot(kind='bar', ax=ax9)
ax9.set_title('IX. Annual Average Signal Ratio', fontsize=14)
ax9.set_ylabel('Mean Ratio')

ax10 = plt.subplot(3, 2, 4)
market_profile = monthly_summary[['hs300_regime_code', 'hs300_target_num', 'hs300_band_width', 'hs300_band_position']]
market_profile.plot(ax=ax10, lw=2)
ax10.set_title('X. Monthly HS300 Regime Variables', fontsize=14)
ax10.axhline(0, color='black', lw=1)

ax11 = plt.subplot(3, 2, 5)
if not stacked_df.empty and len(top_pairs_raw) > 0:
    top_pair = top_pairs_raw.iloc[0]
    x_col = top_pair['factor_1']
    y_col = top_pair['factor_2']
    sns.regplot(
        x=x_col,
        y=y_col,
        data=stacked_df,
        scatter_kws={'alpha': 0.08},
        line_kws={'color': '#d62828'},
        ax=ax11
    )
    ax11.set_title(
        'XI. Strongest Pair Scatter: {} vs {}'.format(
            FACTOR_LABELS.get(x_col, x_col),
            FACTOR_LABELS.get(y_col, y_col),
        ),
        fontsize=14
    )

ax12 = plt.subplot(3, 2, 6)
top_pairs_display = top_pairs_raw.copy()
top_pairs_display['pair'] = top_pairs_display['factor_1'].map(FACTOR_LABELS) + ' vs ' + top_pairs_display['factor_2'].map(FACTOR_LABELS)
top_pairs_display = top_pairs_display.sort_values('abs_corr', ascending=True).tail(8)
ax12.barh(top_pairs_display['pair'], top_pairs_display['abs_corr'], color='#6a994e')
ax12.set_title('XII. Top Absolute Spearman Correlation Pairs', fontsize=14)
ax12.set_xlabel('Average |Correlation|')

plt.tight_layout(rect=[0, 0.03, 1, 0.93])
plt.show()

# ------------------------------------------------------------------------------
# 8. Visualization: Part III - Standalone Heatmap Charts
# ------------------------------------------------------------------------------
# This section outputs dedicated chart-style heatmaps so correlation structure
# can be viewed directly as independent figures in addition to the combined panels.
fig3 = plt.figure(figsize=(26, 10))
plt.suptitle('Factor Correlation Heatmap Report', fontsize=24, fontweight='bold', y=0.98)

ax13 = plt.subplot(1, 3, 1)
plot_heatmap(ax13, avg_spearman, 'I. Stock-Level Average Spearman Heatmap', fmt='.2f')

ax14 = plt.subplot(1, 3, 2)
plot_heatmap(ax14, high_corr_freq, 'II. High-Correlation Frequency Heatmap', fmt='.0%')

ax15 = plt.subplot(1, 3, 3)
plot_heatmap(ax15, market_factor_corr, 'III. HS300 Context Interaction Heatmap', fmt='.2f')

plt.tight_layout(rect=[0, 0.03, 1, 0.93])
plt.show()

fig4 = plt.figure(figsize=(12, 10))
plt.suptitle('Signal-Ratio Correlation Heatmap', fontsize=22, fontweight='bold', y=0.96)
ax16 = plt.subplot(1, 1, 1)
plot_heatmap(ax16, signal_corr, 'IV. Monthly Signal-Ratio Correlation', fmt='.2f')
plt.tight_layout(rect=[0, 0.03, 1, 0.93])
plt.show()

# ------------------------------------------------------------------------------
# 9. Summary Output
# ------------------------------------------------------------------------------
from IPython.display import display
print('\n' + '=' * 110)
print('FINAL SUMMARY: MULTI-FACTOR CORRELATION DIAGNOSTIC')
print('-' * 110)
print('Average Pearson Correlation Matrix:')
display(avg_pearson.round(4))
print('\nAverage Spearman Correlation Matrix:')
display(avg_spearman.round(4))
print('\nFrequency Matrix of |Spearman Corr| >= 0.60:')
display(high_corr_freq.round(4))
print('\nTop Correlation Pairs:')
print(top_pairs_display[['pair', 'corr', 'abs_corr']].sort_values('abs_corr', ascending=False).to_string(index=False))
print('\nHS300 Context vs Monthly Factor Summary Correlation:')
display(market_factor_corr.round(4))
print('=' * 110)
```

## Interpretation of Current Results`nBased on the latest notebook output, the factor library currently shows several important structural features:`n`n- **BBI Slope vs 10D Downtrend: Spearman Corr = 0.7077**`n- **10D Downtrend vs 3D Drawdown: Spearman Corr = 0.5005**`n- **OBV Divergence vs 3D Drawdown: Spearman Corr = 0.4813**`n- **OBV Divergence vs Net Main Inflow: Spearman Corr = -0.4464**`n- **High-correlation frequency of BBI Slope vs 10D Downtrend: 76.42%**`n`n### Practical Reading`nThe most important conclusion is that the factor set is **not fully redundant**, but there is a clearly visible **short-term technical cluster**.
