# 10-Day Downtrend Factor Analysis`n`n## Research Objective`nThis notebook evaluates the predictive value, stability, and practical usefulness of the **10-day downtrend factor** inside the Dynamic Mean Reversion & Multi-Factor Flow Alpha framework. The focus is whether stocks already in a short-term downtrend provide a better oversold-rebound setup in the next monthly holding window.`n

## Factor Definition`nThe factor is built from the same rule used in the strategy code:`n`n- Use the latest **11 trading days** of close-price data`n- Define the raw factor as **latest close / close from 10 trading days earlier - 1**`n- Define the strategy-style signal as:`n  - **Signal**: latest close < close from 10 trading days earlier`n  - **Non-Signal**: latest close >= close from 10 trading days earlier`n`n## Research Hypothesis`nThe working hypothesis is that stocks already under short-term price pressure may be more likely to produce a rebound if they enter the strategy's oversold-candidate set. If the hypothesis holds, the downtrend-signal group should deliver better next-period returns than the non-signal group, and deeper recent weakness should rank slightly better cross-sectionally.`n`n## Data and Test Design`nThe factor is tested on the strategy's fixed 33-stock pool over **2016-01-01 to 2026-04-29**, using monthly cross-sectional snapshots aligned with the reference factor notebooks. The analysis includes rank IC, quintile return comparison, signal-versus-non-signal return comparison, rolling spread inspection, sample-count tracking, regression inspection, and annual heatmap attribution.`n

```python
'''
This diagnostic evaluates the "10-Day Downtrend" factor used by the
Dynamic Mean Reversion & Multi-Factor Flow Alpha strategy.

Core Logic:
1. A stock passes this filter when the latest close is below the close
   from 10 trading days earlier.
2. The script studies whether stocks already in a short-term downtrend
   deliver stronger rebound returns over the next monthly holding window.
3. The research universe is the strategy's fixed 33-stock pool and the
   sampling framework follows the reference JoinQuant factor notebooks.
'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from jqdata import *
from scipy.stats import spearmanr
import warnings

warnings.filterwarnings('ignore')
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['axes.unicode_minus'] = False
sns.set_style("whitegrid")

STOCK_POOL = [
    '601117.XSHG', '601600.XSHG', '601888.XSHG', '300274.XSHE', '300750.XSHE',
    '601919.XSHG', '002049.XSHE', '603881.XSHG', '002335.XSHE', '600089.XSHG',
    '002236.XSHE', '002056.XSHE', '300866.XSHE', '002611.XSHE', '600760.XSHG',
    '300693.XSHE', '002402.XSHE', '002600.XSHE', '300207.XSHE', '603486.XSHG',
    '000591.XSHE', '000027.XSHE', '600011.XSHG', '601899.XSHG', '603799.XSHG',
    '002340.XSHE', '002780.XSHE', '600160.XSHG', '601225.XSHG', '002555.XSHE',
    '600803.XSHG', '300059.XSHE', '002736.XSHE',
]


def safe_qcut(series, groups=5):
    ranked = series.rank(method='first')
    return pd.qcut(ranked, groups, labels=False, duplicates='drop')


def plot_dist(ax, series, color, bins=50, label=None):
    clean = pd.Series(series).dropna()
    if len(clean) == 0:
        return
    if hasattr(sns, 'distplot'):
        sns.distplot(clean, bins=bins, color=color, label=label, ax=ax)
    else:
        ax.hist(clean, bins=bins, color=color, alpha=0.75, label=label)


def calc_icir(ic_series):
    std = ic_series.std()
    if pd.isna(std) or std == 0:
        return np.nan
    return ic_series.mean() / std


# ------------------------------------------------------------------------------
# 1. Sampling Framework
# ------------------------------------------------------------------------------
start_date, end_date = '2016-01-01', '2026-04-29'
all_days = get_trade_days(start_date, end_date)
ts_series = pd.to_datetime(pd.Series(all_days))
rebal_dates = ts_series.groupby([ts_series.dt.year, ts_series.dt.month]).last()
date_list = [d.strftime('%Y-%m-%d') for d in rebal_dates]

dataset = []
print('Initiating 10-Day Downtrend Factor Diagnostic...')

# ------------------------------------------------------------------------------
# 2. Cross-Sectional Data Collection
# ------------------------------------------------------------------------------
for i in range(len(date_list) - 1):
    t0, t1 = date_list[i], date_list[i + 1]
    print('\rProcessing Snapshot: {}'.format(t0), end='')

    hist = get_price(STOCK_POOL, end_date=t0, count=11, fields=['close'], panel=False, fq='pre')
    hold = get_price(STOCK_POOL, start_date=t0, end_date=t1, fields=['close'], panel=False, fq='pre')
    if hist is None or hist.empty or hold is None or hold.empty:
        continue

    for stock in STOCK_POOL:
        h = hist[hist['code'] == stock].sort_values('time')
        p = hold[hold['code'] == stock].sort_values('time')
        if len(h) < 11 or len(p) < 2:
            continue

        factor_raw = h['close'].iloc[-1] / h['close'].iloc[0] - 1.0
        signal = 1 if h['close'].iloc[-1] < h['close'].iloc[0] else 0
        fwd_ret = p['close'].iloc[-1] / p['close'].iloc[0] - 1.0

        dataset.append({
            'date': t0,
            'code': stock,
            'raw': factor_raw,
            'score': -factor_raw,
            'signal': signal,
            'fwd_ret': fwd_ret,
        })

print('\nData collection complete.')

# ------------------------------------------------------------------------------
# 3. Master Dataset Preparation
# ------------------------------------------------------------------------------
master_df = pd.DataFrame(dataset)
master_df['date'] = pd.to_datetime(master_df['date'])
master_df['year'] = master_df['date'].dt.year
master_df['month'] = master_df['date'].dt.month
master_df = master_df.dropna(subset=['score', 'fwd_ret'])

ic_list = []
signal_ret_list = []
group_ret_list = []

for date, group in master_df.groupby('date'):
    if len(group) < 10:
        continue

    ic, _ = spearmanr(group['score'], group['fwd_ret'])
    ic_list.append({'date': date, 'ic': ic})

    group = group.copy()
    if group['score'].nunique() > 1:
        group['quantile'] = safe_qcut(group['score'], 5)
        for q, sub in group.groupby('quantile'):
            group_ret_list.append({
                'date': date,
                'quantile': int(q),
                'ret': sub['fwd_ret'].mean(),
                'count': len(sub),
            })

    pos = group[group['signal'] == 1]['fwd_ret']
    neg = group[group['signal'] == 0]['fwd_ret']
    signal_ret_list.append({
        'date': date,
        'positive_ret': pos.mean() if len(pos) > 0 else np.nan,
        'negative_ret': neg.mean() if len(neg) > 0 else np.nan,
        'spread': (pos.mean() - neg.mean()) if len(pos) > 0 and len(neg) > 0 else np.nan,
        'positive_count': len(pos),
        'negative_count': len(neg),
        'signal_ratio': float(len(pos)) / len(group) if len(group) > 0 else np.nan,
    })

ic_ts = pd.DataFrame(ic_list).set_index('date')
q_ret = pd.DataFrame(group_ret_list).pivot(index='date', columns='quantile', values='ret').sort_index()
sig_ret = pd.DataFrame(signal_ret_list).set_index('date').sort_index()
sig_ret['rolling_spread_12m'] = sig_ret['spread'].rolling(12, min_periods=3).mean()
sig_ret['rolling_positive_ret_12m'] = sig_ret['positive_ret'].rolling(12, min_periods=3).mean()
sig_ret['rolling_negative_ret_12m'] = sig_ret['negative_ret'].rolling(12, min_periods=3).mean()

annual_spread = sig_ret.copy()
annual_spread['year'] = annual_spread.index.year
annual_spread = annual_spread.groupby('year')['spread'].mean()

heat_df = sig_ret.copy()
heat_df['year'] = heat_df.index.year
heat_df['month'] = heat_df.index.month
heat_df = heat_df.pivot_table(index='year', columns='month', values='spread', aggfunc='mean')

# ------------------------------------------------------------------------------
# 4. Visualization: Part I - Signal and Rank Diagnostics
# ------------------------------------------------------------------------------
fig1 = plt.figure(figsize=(22, 22))
plt.suptitle('10-Day Downtrend Factor Diagnostic: Part I', fontsize=24, fontweight='bold', y=0.96)

ax1 = plt.subplot(3, 2, 1)
(1 + sig_ret[['positive_ret', 'negative_ret', 'spread']].fillna(0)).cumprod().plot(ax=ax1, lw=2)
ax1.set_title('I. Cumulative Return of Signal / Non-Signal / Spread', fontsize=14)

ax2 = plt.subplot(3, 2, 2)
if len(ic_ts) > 0:
    ic_ts['ic'].fillna(0).cumsum().plot(ax=ax2, color='#1d3557', lw=2)
ax2.axhline(0, color='black', lw=1)
ax2.set_title('II. Cumulative Rank IC', fontsize=14)

ax3 = plt.subplot(3, 2, 3)
if len(q_ret) > 0:
    (1 + q_ret.fillna(0)).cumprod().plot(ax=ax3, lw=2)
ax3.set_title('III. Cumulative Return by Downtrend Severity Quintile', fontsize=14)

ax4 = plt.subplot(3, 2, 4)
sig_ret['rolling_spread_12m'].plot(ax=ax4, color='#e76f51', lw=2)
ax4.axhline(0, color='black', lw=1)
ax4.set_title('IV. 12-Month Rolling Mean Spread', fontsize=14)

ax5 = plt.subplot(3, 2, 5)
sig_ret[['positive_count', 'negative_count']].plot(ax=ax5, lw=2, color=['#2a9d8f', '#adb5bd'])
ax5.set_title('V. Monthly Signal and Non-Signal Count', fontsize=14)
ax5.set_ylabel('Sample Size')

ax6 = plt.subplot(3, 2, 6)
annual_spread.plot(kind='bar', color='#2a9d8f', ax=ax6)
ax6.axhline(0, color='black', lw=1)
ax6.set_title('VI. Annual Mean Spread Return', fontsize=14)

plt.tight_layout(rect=[0, 0.03, 1, 0.93])
plt.show()

# ------------------------------------------------------------------------------
# 5. Visualization: Part II - Distribution and Attribution
# ------------------------------------------------------------------------------
fig2 = plt.figure(figsize=(22, 24))
plt.suptitle('10-Day Downtrend Factor Diagnostic: Part II', fontsize=24, fontweight='bold', y=0.96)

ax7 = plt.subplot(3, 2, 1)
plot_dist(ax7, master_df[master_df['signal'] == 1]['fwd_ret'], color='#2a9d8f', bins=50, label='Signal')
plot_dist(ax7, master_df[master_df['signal'] == 0]['fwd_ret'], color='#e63946', bins=50, label='Non-Signal')
ax7.axvline(0, color='black', lw=1)
ax7.legend()
ax7.set_title('VII. Forward Return Distribution by Signal State', fontsize=14)

ax8 = plt.subplot(3, 2, 2)
sample_df = master_df.sample(n=min(len(master_df), 3000), random_state=42)
sns.regplot(x='raw', y='fwd_ret', data=sample_df, scatter_kws={'alpha': 0.10}, line_kws={'color': '#d62828'}, ax=ax8)
ax8.set_title('VIII. 10-Day Return vs Forward Return', fontsize=14)

ax9 = plt.subplot(3, 2, 3)
sns.boxplot(x='signal', y='fwd_ret', data=master_df, palette=['#e63946', '#2a9d8f'], ax=ax9)
ax9.set_title('IX. Forward Return Boxplot by Signal State', fontsize=14)
ax9.set_xlabel('Signal')

ax10 = plt.subplot(3, 2, 4)
plot_dist(ax10, master_df['raw'], color='#457b9d', bins=60)
ax10.axvline(0, color='red', linestyle='--', lw=1.5)
ax10.set_title('X. Distribution of 10-Day Price Change', fontsize=14)

ax11 = plt.subplot(3, 2, 5)
sig_ret[['rolling_positive_ret_12m', 'rolling_negative_ret_12m']].plot(ax=ax11, lw=2, color=['#2a9d8f', '#6c757d'])
ax11.axhline(0, color='black', lw=1)
ax11.set_title('XI. 12-Month Rolling Mean Return by Group', fontsize=14)

ax12 = plt.subplot(3, 2, 6)
sns.heatmap(heat_df, annot=True, fmt='.2%', cmap='RdYlGn', center=0, ax=ax12)
ax12.set_title('XII. Monthly Spread Heatmap by Year', fontsize=14)
ax12.set_xlabel('Month')

plt.tight_layout(rect=[0, 0.03, 1, 0.93])
plt.show()

# ------------------------------------------------------------------------------
# 6. Summary Output
# ------------------------------------------------------------------------------
print('\n' + '=' * 90)
print('FINAL SUMMARY: 10-DAY DOWNTREND FACTOR')
print('-' * 90)
print('Average Rank IC: {:.4f}'.format(ic_ts['ic'].mean()))
print('ICIR: {:.4f}'.format(calc_icir(ic_ts['ic'])))
print('Signal Mean Return: {:.2f}%'.format(sig_ret['positive_ret'].mean() * 100))
print('Non-Signal Mean Return: {:.2f}%'.format(sig_ret['negative_ret'].mean() * 100))
print('Signal Spread Mean Return: {:.2f}%'.format(sig_ret['spread'].mean() * 100))
print('=' * 90)
```

However, the advantage is only **0.06%**, which is too small to support use as a decisive standalone stock-selection factor.`n`n### Strategic Relevance`nIn practical terms, this factor looks more suitable as a **background-condition or pool-entry filter** than as a return-ranking engine. It helps define whether a stock is already in a short-term weakening state, which fits the strategy's oversold-rebound philosophy, but it adds only limited incremental forecasting power by itself.`n`n### Research Conclusion`nAt the current stage, the 10-day downtrend factor should be interpreted as a **supporting oversold-context factor** rather than a strong alpha factor. Its main value likely comes from working together with money-flow confirmation, short-term drawdown ranking, and broader market-regime control, where it helps define the right type of rebound setup instead of independently driving final selection.`n
